```python
import typing
import matplotlib.pyplot as plt
from sympy import Union, Interval, Complement, EmptySet, oo, S
import numpy as np
import ipywidgets as widgets


def logics(a, b):
    premise = [(-6, -2), (5, 8)]
    replacement: dict[typing.Union[str, int], int] = {"-oo": -10, "+oo": 10}
    # Replace oo with 10
    tempPredicate = []
    for couple in premise:
        new_couple: list[typing.Union[int, str]] = [
            replacement.get(n, n) for n in couple
        ]
        tempPredicate.append(tuple(new_couple))
    premise = tempPredicate
    # Create premise plot
    plotPremise = []
    for couple in premise:
        couple = list(couple)
        couple[1] = couple[1] - couple[0]
        plotPremise.append(tuple(couple))
    # Build Sets
    pSet = EmptySet
    for couple in premise:
        pSet = Union(pSet, Interval(couple[0], couple[1]))
    pSetC = pSet.complement(Interval(-10, 10))
    eSet = Interval(a, b)
    # Create complement plot
    plotComplement = []
    for interval in pSetC.args:
        plotComplement.append((interval.start, interval.end - interval.start))
    # What do we have here?
    label = ""
    if pSet.intersect(eSet) == EmptySet:
        label = "Strong Speculation"
    else:
        if pSetC.intersect(eSet) == EmptySet:
            label = "Hypothesis"
        else:
            if pSet.intersect(eSet) == pSet:
                label = "Consequence"
            else:
                label = "Weak Speculation"
    fig, gnt = plt.subplots(figsize=(8, 2))
    gnt.set_yticks([0.5, 1.5, 2.5])
    gnt.set_xlim(-10, 10)
    gnt.set_ylim(0, 3)
    gnt.set_yticklabels([label, "Complement", "Premise"])
    labels = list(range(-10, 11, 1))
    gnt.set_xticks(labels)
    str_labels = [str(x) for x in labels]
    str_labels[0] = "-∞"
    str_labels[20] = "+∞"
    gnt.set_xticklabels(str_labels)
    gnt.broken_barh(plotPremise, (2, 1), facecolors=("tab:green"))
    gnt.broken_barh(plotComplement, (1, 1), facecolors=("tab:red"))
    if a < b:
        gnt.broken_barh([(a, b - a)], (0, 1), facecolors=("tab:orange"))


widgets.interact(
    logics,
    a=widgets.FloatSlider(min=-9, max=9, step=1, value=-1),
    b=widgets.FloatSlider(min=-9, max=9, step=1, value=1),
)
```
